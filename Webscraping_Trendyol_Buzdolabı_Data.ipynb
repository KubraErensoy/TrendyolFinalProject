```python
def getAndParseURL(url):
    result = requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
def getCommentPageHTML(url):
    driver.get(url)
    SCROLL_PAUSE_TIME = 0.5

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    defhtml = driver.page_source
    defsoup = BeautifulSoup(defhtml, "html.parser")
    return defsoup  
```

```python
def getStar(url):
    star = url.find("div",{"class":"pr-rnr-sm-p"}).span.text
    return float(star)
```

```python
SAYFA_NO = []

def find_sayfa_no():
    SAYFA_NO.clear()
    html = getAndParseURL("https://www.trendyol.com/buzdolabi-x-c103623?")
    for i in range(1,43):
        SAYFA_NO.append("https://www.trendyol.com/buzdolabi-x-c103623?" + "&pi=" + str(i))
    return SAYFA_NO

find_sayfa_no()
        
   
```

```python
TUM_URUNLER = []

def tum_urunler():
    TUM_URUNLER.clear()
    a=0
    for i in SAYFA_NO:
        html = getAndParseURL(i)
        for sonuc in html.findAll("div",{"class":"p-card-wrppr"}):
            TUM_URUNLER.append("https://www.trendyol.com" + sonuc.find("div",{"class":"p-card-chldrn-cntnr"}).a["href"])
            a = a+1
    print("Tum Urunler",a)          
    return TUM_URUNLER
tum_urunler()
```
